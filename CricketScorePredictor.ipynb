```python
# 8. Adding batting team strength based on past performance (mean runs in past matches)
df['batting_team_strength'] = df.groupby('batting_team')['runs'].transform('mean')

# 9. Adding bowling team strength (mean wickets taken in past matches)
df['bowling_team_strength'] = df.groupby('bowling_team')['runs'].transform('mean')
# 10. Adding venue impact (average runs scored in past matches at the same venue)
df['venue_avg_runs'] = df.groupby('venue')['runs'].transform('mean')

# 11. Batting impact (difference between team performance and venue average)
df['batting_impact'] = df['current'] - df['venue_avg_runs']
 #13. Adding match_id-specific features
df['match_venue'] = df['match_id'].astype(str) + '_' + df['venue']

# 15. Bowling team wickets in last 5 overs
df['bowling_team_last_5_wickets'] = df.groupby('bowling_team')['player_dismissed'].transform(lambda x: x.rolling(window=30).sum())

# Feature that combines the runs and wickets left for the team
df['runs_wickets_left_ratio'] = df['runs'] / (df['wickets_left'] + 1)  # Avoid division by zero
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from catboost import CatBoostRegressor
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.impute import SimpleImputer
import pandas as pd

transformer = ColumnTransformer(
    transformers=[('transformer', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])],
    remainder='passthrough'
)

# Defining models for testing
models = {
    'XGBoost': XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1),
    'Random Forest': RandomForestRegressor(n_estimators=1000, random_state=1),
    'CatBoost': CatBoostRegressor(iterations=1000, depth=10, learning_rate=0.1, random_seed=1, silent=True),
    'Linear Regression': LinearRegression(),
    'Lasso Regression': Lasso(alpha=0.1)
}

# Pipeline setup with ColumnTransformer and models
for model_name, model in models.items():
    print(f"Training {model_name}...")
    
    pipe = Pipeline(steps=[
        ('step1', transformer),
        ('step2', SimpleImputer(strategy='mean')),  # Handle missing values
        ('step3', StandardScaler()),
        ('step4', model)
    ])
    
    # Fitting the model
    pipe.fit(X_train, y_train)
    
    # Making predictions
    y_pred = pipe.predict(X_test)
    
    # Evaluating the model
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    
    print(f"{model_name} - R2 Score: {r2}")
    print(f"{model_name} - Mean Absolute Error: {mae}\n")

```
